# 02_EDA – Exploración de Datos
Análisis exploratorio para evaluar el impacto del riego en la productividad agrícola.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

data_path = '/mnt/data'
files = os.listdir(data_path)
files
```

## Cargar un dataset procesado

```python
example_file = 0
for f in files:
    if f.endswith('.csv'):
        example_file = os.path.join(data_path, f)
        break
example_file
```

```python
if example_file:
    df = pd.read_csv(example_file)
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    df.head()
else:
    'No CSV files found.'
```

## Estadísticos descriptivos

```python
df.describe(include='all')
```

## Distribución del rendimiento

```python
if 'rendimiento' in df.columns:
    plt.hist(df['rendimiento'].dropna())
    plt.title('Distribución del rendimiento')
    plt.xlabel('Rendimiento')
    plt.ylabel('Frecuencia')
    plt.show()
else:
    'No existe la columna rendimiento.'
```

## Comparación riego vs temporal

```python
if {'modalidad_hidrica','rendimiento'}.issubset(df.columns):
    df.groupby('modalidad_hidrica')['rendimiento'].mean()
else:
    'No existen columnas suficientes.'
```

## Gráfico riego vs rendimiento

```python
if {'modalidad_hidrica','rendimiento'}.issubset(df.columns):
    grouped = df.groupby('modalidad_hidrica')['rendimiento'].mean()
    plt.bar(grouped.index.astype(str), grouped.values)
    plt.title('Rendimiento promedio por modalidad hídrica')
    plt.xlabel('Modalidad')
    plt.ylabel('Rendimiento')
    plt.show()
else:
    'No existen columnas suficientes.'
```
