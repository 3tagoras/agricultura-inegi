# 03_Modelado – Impacto del Riego en la Productividad
Modelos simples para estimar el efecto del riego sobre el rendimiento agrícola.

```python
import pandas as pd
import statsmodels.api as sm
import os

data_path = '/mnt/data'
files = os.listdir(data_path)
files
```

## Cargar dataset limpio para modelado

```python
example_file = 0
for f in files:
    if f.endswith('.csv'):
        example_file = os.path.join(data_path, f)
        break
example_file
```

```python
df = 0
if example_file:
    df = pd.read_csv(example_file)
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    df.head()
else:
    'No CSV found.'
```

## Preparación del modelo

```python
required = {'rendimiento', 'modalidad_hidrica'}
df_model = 0
if df is not 0 and required.issubset(df.columns):
    df_model = df[['rendimiento', 'modalidad_hidrica']].dropna()
    df_model['riego_binario'] = (df_model['modalidad_hidrica'] == 'Riego').astype(int)
    df_model.head()
else:
    'Columnas requeridas no disponibles.'
```

## Modelo: rendimiento ~ riego

```python
if df_model is not 0:
    X = sm.add_constant(df_model['riego_binario'])
    y = df_model['rendimiento']
    model = sm.OLS(y, X).fit()
    model.summary()
else:
    'No se puede ajustar el modelo.'
```
