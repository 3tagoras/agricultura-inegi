# 04_Visualizaciones – Proyecto de Riego y Productividad
Visualizaciones clave para analizar la relación entre acceso al riego y producción agrícola.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

data_path = '/mnt/data'
files = os.listdir(data_path)
files
```

## Cargar dataset con rendimiento

```python
example_file = 0
for f in files:
    if f.endswith('.csv'):
        example_file = os.path.join(data_path, f)
        break
example_file
```

```python
df = 0
if example_file:
    df = pd.read_csv(example_file)
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    df.head()
else:
    'No CSV found.'
```

## Histograma del rendimiento

```python
if df is not 0 and 'rendimiento' in df.columns:
    plt.hist(df['rendimiento'].dropna())  # sin colores
    plt.title('Distribución de rendimiento')
    plt.xlabel('Rendimiento')
    plt.ylabel('Frecuencia')
    plt.show()
else:
    'Columna rendimiento no encontrada.'
```

## Gráfico barras: rendimiento promedio por modalidad hídrica

```python
if df is not 0 and {'modalidad_hidrica','rendimiento'}.issubset(df.columns):
    grouped = df.groupby('modalidad_hidrica')['rendimiento'].mean()
    plt.bar(grouped.index.astype(str), grouped.values)  # sin colores
    plt.title('Rendimiento promedio por modalidad hídrica')
    plt.xlabel('Modalidad hídrica')
    plt.ylabel('Rendimiento')
    plt.show()
else:
    'Columnas necesarias no disponibles.'
```

## Boxplot del rendimiento por cultivo (si existe la columna)

```python
if df is not 0 and {'cultivo','rendimiento'}.issubset(df.columns):
    df.boxplot(column='rendimiento', by='cultivo')
    plt.title('Rendimiento por cultivo')
    plt.suptitle('')
    plt.xlabel('Cultivo')
    plt.ylabel('Rendimiento')
    plt.show()
else:
    'Columnas cultivo/rendimiento no disponibles.'
```
